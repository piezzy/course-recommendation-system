```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('udemy_course_data.csv')
df.head()
```

```python
df.info()
```

EDA on Subject

```python
len(df['course_title'].unique())
```

```python
plt.figure(figsize=(10, 6))
plt.title('Number of Courses per Subject')
sn.countplot(df['subject'], palette='Set2')
```

```python
plt.figure(figsize=(10, 6))
df['subject'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sn.color_palette('Set2'),  title='Course Distribution by Subject (%)')
```

EDA on Subs

```python
df.groupby('subject')['num_subscribers'].count().plot(kind='barh', figsize=(10, 6), color=sn.color_palette('Set2'), title='Number of Courses per Subject')
```

```python
df.groupby('subject')['num_subscribers'].sum().plot(kind='barh', figsize=(10, 6), color=sn.color_palette('Set2'), title='Total Subscribers per Subject')
```

```python
df.groupby('subject')['num_subscribers'].sum().plot(kind='pie', figsize=(10, 6), color=sn.color_palette('Set2'), title='Total Subscribers per Subject')
```

```python
print(df.groupby('subject')['num_subscribers'].sum())
print(df.groupby('subject')['num_subscribers'].count())
```
